```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #just another library for visualization
from matplotlib.ticker import PercentFormatter
sns.set()
```

```python
# Load real estate data set for the histogram
df_real_estate = pd.read_csv("histogram_data.csv")
```

```python
df_real_estate.head(20)
```

```python
plt.hist(df_real_estate["Price"])
plt.show() #the most basic style
```

```python
# override the default matplotlib style, to avoid the gray background and grid
sns.set_style("white") 
plt.figure(figsize = (8, 6)) # determine the size of the figure
plt.hist(df_real_estate["Price"], # the variable on which to create the histogram
         bins = 9, # create a histogram with 8 bins
         #you are allowed to alter bins above 
         color = "#FF5733")
plt.title("Distribution of Real Estate Prices based on Rubles", fontsize = 25, weight = "bold")
plt.xlabel("Price in (00's Rubles)",fontsize=20)  #its january now, Crypto is going down
plt.ylabel("Number of Properties",fontsize=20)
sns.despine()
 # removes the top and right border of our graph
plt.show()
#its a right skewed distribution
```
