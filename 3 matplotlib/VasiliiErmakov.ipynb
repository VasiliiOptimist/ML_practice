```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default

from matplotlib.ticker import PercentFormatter #converts values into percentage format
```

<h1>BARCHART</h1>

```python
df_used_cars = pd.read_csv("BARCHART/bar_chart_data.csv")
df_sorted_used_cars = df_used_cars.sort_values('Cars Listings', ascending=True)
df_sorted_used_cars
```

```python
def add_numbers(brands, number_of_listings):
        for i in range(len(brands)):
                plt.text(number_of_listings[i]//2, brands[i], number_of_listings[i])
```

```python
brands = df_sorted_used_cars['Brand']
listings = df_sorted_used_cars['Cars Listings']

fig, ax = plt.subplots(figsize=(8, 10))
ax.barh(y=brands,
        width=listings,
        color='y')
fig.set_facecolor('#fffcdc')
ax.set_facecolor('#eaffe5')

add_numbers(brands, listings)

ax.set_ylabel('Brands', weight='bold', fontsize=16)
ax.tick_params(axis='y', labelrotation=45)
ax.set_xlabel('Number of Listings', weight='bold', fontsize=16)
ax.set_title('Used Cars Listings', fontsize=20, weight='heavy')
plt.show()
```

<h1>HISTOGRAM</h1>

```python
df_survey_data = pd.read_csv("HISTOGRAM/histogram_survey_data.csv")
df_survey_data
```

```python
sns.set_style('ticks')
fig, ax = plt.subplots(figsize = (8, 6))

ax.hist(df_survey_data["Age"],
        bins = 10,
        color = 'orange')
ax.set_ylabel('Number of Participants', fontsize = 16)
ax.set_xlabel('Age', fontsize = 16)
ax.tick_params(axis='x', width=3, labelsize='large', color='black')
ax1 = ax.twinx()
ax1.plot(sorted(df_survey_data['Age'].unique()),
        df_survey_data.groupby('Age')['Income'].mean(),
        color='red',
        marker='s')
ax1.set_ylabel('Income', fontsize=16, color='red')
ax1.tick_params(axis='y', colors='red')
sns.despine()
plt.show()
```

<b>Insights:</b>
1. Getting older is profitable=)
2. The participants seems to be employees in the factory, because there are many middle-age people and few old people.

<h1>LINECHART</h1>

```python
df_returns1020 = pd.read_csv('LINE CHART/returns1020.csv')
df_returns1020['new_date'] = pd.to_datetime(df_returns1020['Date']) 
df_returns1920 = df_returns1020[(df_returns1020.new_date > '2019-02-01') &
                                (df_returns1020.new_date < '2019-05-01')]
```

```python
df_returns1020
```

```python
plt.figure(figsize=(15, 8))

plt.plot(df_returns1920['new_date'], df_returns1920['GSPCRet'], color='red')
plt.plot(df_returns1920['new_date'], df_returns1920['FTSERet'], color='black')
plt.title('GSPCRet and FTSERet Returns (2019.02-2019.05)', fontsize=20, fontweight='heavy')
plt.ylabel('Returns', fontsize=20, fontweight='bold')
plt.xlabel('Date', fontsize=20, fontweight='bold')
plt.legend(labels=['GSPCRet', 'FTSERet'], fontsize='large')
sns.despine()
plt.show()
```

<h1>STACKED AREA CHART</h1>

```python
df_area_chart = pd.read_csv("STACKED AREA CHART/area_chart_homework_data.csv")
```

```python
df_area_chart['new_date'] = pd.to_datetime(df_area_chart['Date'])
df_area_chart
```

```python
labels = ['International', 'Domestic', 'Services']

sns.set_style('whitegrid')
plt.figure(figsize=(16, 9))
plt.stackplot(df_area_chart['new_date'],
             df_area_chart['International'],
             df_area_chart['Domestic'],
             df_area_chart['Services'],
             colors=['#FF2A00', '#FF7500', '#FFA800'],
             edgecolor='none')
plt.xticks(df_area_chart['new_date'], rotation=45, fontsize=14)
plt.legend(labels = labels, loc='upper left', fontsize=14)
plt.ylabel('Revenue by category', fontsize=18, weight='bold')
plt.xlabel('Months', fontsize=18, weight='bold')
plt.title('Revenue of the Printer Company (2020)', fontsize=20, weight='bold')
sns.despine()
plt.show()
```

The chart shows the revenue of the company in 2020.
The most important part of the revenue is the international part.
There is a huge drop from April to August, but a quick recovery after.
