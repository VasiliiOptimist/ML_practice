```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default

from matplotlib.ticker import PercentFormatter #converts values into percentage format
```

```python
# Read the car data csv file for this task.
# Make sure the file is in the same directory, as your Python Notebook, otherwise you'll need to specify the file location.
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
plt.figure(figsize = (20, 10)) #determine the size of the chart
# To create a bar chart with matplotlib you can use the 'bar' function.
plt.bar(x = df_used_cars["Brand"], #specify the x axis
        height = df_used_cars["Cars Listings"], #specify the y axis
        color = "pink") # the color for the bars

plt.xticks(rotation = 45,fontsize = 20) # rotate and format the labels for the x-axis
#rotation can be useful for you to fit the names of objects
plt.yticks(rotation = 45, fontsize = 20) # format the y-axis
plt.title("Cars Listings by Viktor", fontsize = 50, fontweight = "bold") #add and format the title for the chart
plt.ylabel("Number of Listings", fontsize = 20 ) #add a title for the y-axis
plt.savefig("Used Cars Bar.png") # you can export your chart as a picture
plt.show() # depending on which environment you're using without this line you're chart may not display properly.
```

```python
df_sorted_used_cars = df_used_cars.sort_values('Cars Listings', ascending=True)
df_sorted_used_cars
```

```python
def add_numbers(brands, number_of_listings):
        for i in range(len(brands)):
                plt.text(number_of_listings[i]//2, brands[i], number_of_listings[i])

brands = df_sorted_used_cars['Brand']
listings = df_sorted_used_cars['Cars Listings']

fig, ax = plt.subplots(figsize=(8, 10))
ax.barh(y=brands,
        width=listings,
        color='y')
fig.set_facecolor('#fffcdc')
ax.set_facecolor('#eaffe5')

add_numbers(brands, listings)

ax.set_ylabel('Brands', weight='bold', fontsize=16)
ax.tick_params(axis='y', labelrotation=45)
ax.set_xlabel('Number of Listings', weight='bold', fontsize=16)
ax.set_title('Used Cars Listings', fontsize=20, weight='heavy')
plt.show()
```
