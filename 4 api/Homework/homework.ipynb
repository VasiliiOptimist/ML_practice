```python
#Impoting libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#BarChart for Steam game tags
from matplotlib.ticker import PercentFormatter
steam_tags = pd.read_csv('steam_game_tags.csv')

colour_list =["#0b5982", "#145784", "#1d5585", "#255386", "#2e5187",\
              "#364e87", "#3f4b86", "#474885", "#4f4583", "#564180",\
              "#5d3d7d", "#64397a", "#6b3475", "#712f70", "#772a6a",\
              "#7b2464", "#801e5e", "#841756", "#870f4f", "#890547"]

sns.set_style('darkgrid')
plt.figure(figsize = (12, 8))
plt.barh(y = steam_tags['Tag'],
         width = steam_tags ['Amount'],
         color = colour_list )

plt.yticks(rotation = 0, fontsize = 11)
plt.xticks(rotation = 50, fontsize = 11)
plt.title('Amount of Steam game tags', fontsize = 20)
plt.ylabel ('Steam game tags', fontsize = 11)
plt.xlabel ('Amount of games', fontsize = 11)
plt.savefig ('Steam_game_tags.png')
plt.show()
plt.rcdefaults()
```

```python
#Histogram of Steam sales prices
steam_sale_prices = pd.read_csv('steam_game_sale_prices.csv')

sns.set_style('white')
xticks = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950, 1000]
plt.figure(figsize = (10,8))
plt.hist(steam_sale_prices['Price'],
        bins = 20,
        color = 'green')
plt.title('Distribution of Steam games prices during sale', fontsize = 25)
plt.xlabel('Price in Roubles ↑', fontsize = 25)
plt.ylabel('Amount of Games ↓', rotation = 90, fontsize =25)
plt.annotate(
    'The price with the most abundant number of sale games',
    xy=(50, 380), arrowprops=dict(arrowstyle='->', color ='k'), xytext=(200, 145), fontsize = 15, rotation = -31)
plt.xticks(xticks, rotation = 90)
sns.despine()

plt.show()
plt.rcdefaults()
```

```python
#Scatter Plot for Steam sale prices, game ratings nad price cuts
steam_sale_prices = pd.read_csv('steam_game_sale_prices.csv')

sns.set_style('whitegrid')
plt.figure(figsize = (10, 6))
scatter = plt.scatter(steam_sale_prices['Price'],
                      steam_sale_prices['Rating'],
                     c = steam_sale_prices['Price cut'],
                     cmap = 'magma',
                     alpha = 0.5,
                     marker = 'x')

plt.legend (*scatter.legend_elements(),
           loc = 'lower right',
           title = 'Price cut')

plt.title('Relationship between Steam game price, rating and price cuts', fontsize = 20)
plt.xlabel('Game price', fontsize = 20)
plt.ylabel('Game rating', fontsize = 20)
plt.show()
plt.rcdefaults()
```

```python
#Linechart of Dota 2 playerbase changes
dota_playerbase = pd.read_csv('dota_playerbase.csv')
plt.xkcd()

sns.set_style('whitegrid')
dota_playerbase['iso_date'] = pd.to_datetime(dota_playerbase['Month'], dayfirst=True)
plt.figure(figsize = (10, 4))
plt.plot(dota_playerbase['iso_date'], dota_playerbase['Peak Players'], color = 'k')
plt.rcParams['lines.linestyle'] = '--'
plt.plot(dota_playerbase['iso_date'], dota_playerbase['Avg. Players'], color = 'k')
plt.title("Number of Dota 2 players")
plt.ylabel('Millions of players')
plt.xlabel('Years')
plt.legend(labels = ['Peak players','Average players'], fontsize = 'medium')
plt.show()
plt.rcdefaults()
```

```python
#Stacked area chart for the most popular games in Steam
top_games = pd.read_csv('top_played_games_on_steam.csv')
top_games['iso_date'] = pd.to_datetime(top_games['Date'], dayfirst=True)

sns.set(rc={'axes.facecolor':'lightblue'})
colour_list_st = ['#FFD46E','#b39f00','#73b96c','#6ec2b5','#adc1c8']
games = ['Team Fortress 2','Grand Theft Auto V', 'Counter-Strike: Global Offensive',\
         'Dota 2', 'PUBG: BATTLEGROUNDS']
plt.figure(figsize = (10, 4))
plt.stackplot(top_games['Date'],
              top_games['Team Fortress 2'],
              top_games['Grand Theft Auto V'],
              top_games['Counter-Strike: Global Offensive'],
              top_games['Dota 2'],
          #    top_games['Cyberpunk 2077'],
          #    top_games['Apex Legends'],
          #    top_games['Terraria'],
              top_games['PUBG: BATTLEGROUNDS'],
              colors=colour_list_st, edgecolor = 'k'
             )

plt.legend(labels = games, loc = 'upper left', fontsize ='small')
plt.title('Total Steam playerbase from the most popular games', weight ='bold')
plt.ylabel('Millions of players')
plt.xticks([1, 7, 19, 31, 43, 55, 67, 79, 91, 103, 115],\
           ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022'])
plt.show()
plt.rcdefaults()
#P.S.Yeah, I cheated.
#P.P.S. Nope, I fixed it.
```
